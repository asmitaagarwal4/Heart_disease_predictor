```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
HDNames= ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','hal','HeartDisease']
Data = pd.read_excel('Ch3.ClevelandData.xlsx', names=HDNames)
```

```python
print(Data.head(20))
print(Data.info())
summary = Data.describe()
print(summary)
```

```python
import numpy as np
```

```python
DataNew = Data.replace('?', np.nan)
```

```python
print(DataNew.info())
```

```python
print(DataNew.describe())
```

```python
print(DataNew.isnull().sum())
```

```python
DataNew = DataNew.dropna()
```

```python
print(DataNew.info())
```

```python
print(DataNew.isnull().sum())
```

```python
InputNames = HDNames
InputNames.pop()
```

```python
Input = pd.DataFrame(DataNew.iloc[:, 0:13],columns=InputNames)
```

```python
Target = pd.DataFrame(DataNew.iloc[:, 13],columns=['HeartDisease'])
```

```python
fig, ax= plt.subplots(figsize=(15,15))
sns.boxplot(data= Input, ax=ax)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
print(scaler.fit(Input))
```

```python
InputScaled = scaler.fit_transform(Input)

InputScaled = pd.DataFrame(InputScaled,columns=InputNames)

summary = InputScaled.describe()
summary = summary.transpose()
print(summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax= plt.subplots(figsize=(15,15))
sns.boxplot(data= InputScaled, ax=ax)
#plt.savefig("Boxplot.jpg")
```

```python
fig, ax= plt.subplots(figsize=(15,15))
boxplot = InputScaled.boxplot(column=InputNames,showmeans=True)
plt.show()
```

```python
pd.plotting.scatter_matrix(InputScaled, figsize=(15, 15))
plt.show()
```

```python
CorData = InputScaled.corr(method='pearson')
```

```python
sns.heatmap(CorData)
```

```python
#Split the data
from sklearn.model_selection import train_test_split

Input_train, Input_test, Target_train, Target_test = train_test_split(InputScaled, Target, test_size = 0.30, random_state = 5)
print(Input_train.shape)
print(Input_test.shape)
print(Target_train.shape)
print(Target_test.shape)
```

```python
Target_test.value_counts()
```

## Data Imbalancing Technique- SMOTE

```python
from imblearn.over_sampling import SMOTE
smote= SMOTE(random_state=42)
Input_train_resampled, Target_train_resampled=smote.fit_resample(Input_train, Target_train)

print("\nResampled Class Distribution:")
(Target_train_resampled).value_counts()
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classification=LogisticRegression()
classification.fit(Input_train_resampled, Target_train_resampled)
```

```python
Y_predictions= classification.predict(Input_test)
Y_predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Target_test, Y_predictions)
```

```python
from sklearn.metrics import classification_report
target_names=['No Heart Disease', 'Having Heart Disease']
print(classification_report(Target_test, Y_predictions, target_names=target_names))
```

```python
import pickle
pickle.dump(classification, open("HD_Logistic_model.pkl", "wb"))
model=pickle.load(open("HD_Logistic_model.pkl", "rb"))
```

```python
model.predict(Input_test)
```

# Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier()
model.fit(Input_train_resampled,Target_train_resampled)
```

```python
Y_pred= model.predict(Input_test)
```

```python
Y_pred
```

```python
from sklearn import tree
plt.figure(figsize=(20, 20))
tree.plot_tree(model, filled=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Target_test, Y_pred)
```

```python
from sklearn.metrics import classification_report
target_names=['No Heart Disease', 'Having Heart Disease']
print(classification_report(Target_test, Y_pred, target_names=target_names))
```

# Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(Input_train_resampled, Target_train_resampled)
```

```python
rf.score(Input_test, Target_test)
```

### Hyperparameter Tuning - GridSearchCV

```python
grid_param= {
    "n_estimators": [50,100,120],
    "criterion":["gini", "entropy"],
    "max_depth": range(5),
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search_rf= GridSearchCV(param_grid= grid_param, cv=10, n_jobs=6, verbose=1, estimator=rf)
grid_search_rf.fit(Input_train_resampled, Target_train_resampled)
```

```python
grid_search_rf.best_params_
```

```python
rf_new= RandomForestClassifier(n_estimators= 50, criterion= 'gini', max_depth= 2)
rf_new.fit(Input_train_resampled, Target_train_resampled)
```

```python
rf_new.score(Input_test, Target_test)
```
